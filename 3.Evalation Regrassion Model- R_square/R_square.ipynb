```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("data.csv",sep=";",header=None)
x = df.iloc[:,0].values.reshape(-1,1)
y = df.iloc[:,1].values.reshape(-1,1)
```

```python
rf = RandomForestRegressor(n_estimators=100,random_state=42) 
#100 => kaç tane Decission Tree kullanacaksın, 42 => n sayıda random bölme kod tekrar çalıştığında aynı sonucu verir
rf.fit(x,y) 
```

```python
y_head = rf.predict(x)
from sklearn.metrics import r2_score
print ("r_score:",r2_score(y,y_head)) # gerçek y değerleri ve predict edilen y_head değerlerdir
```

```python
# Lineer REg
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x,y)
```

```python
y_head = lr.predict(x)
```

```python
from sklearn.metrics import r2_score

print("r_score",r2_score(y,y_head))
```
