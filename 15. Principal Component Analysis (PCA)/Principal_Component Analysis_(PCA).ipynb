```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = iris.data
feature_name =iris.feature_names
y = iris.target
```

```python
df = pd.DataFrame(data,columns=feature_name)
df["sinif"] = y
```

```python
x =data
from sklearn.decomposition import PCA
pca = PCA(n_components=2,whiten=True) # whitten True olursa normalide etmek demek 0-1 etmek


```

```python
pca.fit(x)
x_pca = pca.transform(x)
print("veriance ratio",pca.explained_variance_ratio_) # büyük olan birinci pca

print("sum",sum(pca.explained_variance_ratio_)) # yüzde 97 oranında datayı korumuşum demek

```

```python
df["p1"]=x_pca[:,0]
df["p2"] =x_pca[:,1]
```

```python
#veri görselleştirme
color =["red","green","blue"]
for i in range(3):
    plt.scatter(df.p1[df.sinif == i],df.p2[df.sinif ==i],color=color[i],label = iris.feature_names[i])
plt.legend()
plt.show()
```
