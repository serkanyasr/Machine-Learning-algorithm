```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
#dataseti oluşturma  K değeri 3 np ile data oluşturma

x1 = np.random.normal(25,5,1000) # Gaussiyum dağılımı 25-5 arasında değerler üren gaussi değeri 100 olsun
y1 = np.random.normal(25,5,1000)

x2 = np.random.normal(55,5,1000) # Gaussiyum dağılımı 25-5 arasında değerler üren gaussi değeri 100 olsun
y2 = np.random.normal(60,5,1000)


x3 = np.random.normal(55,5,1000) # Gaussiyum dağılımı 25-5 arasında değerler üren gaussi değeri 100 olsun
y3 = np.random.normal(15,5,1000)
```

```python
x = np.concatenate((x1,x2,x3),axis=0) #yukarıdan aşağıya birleştir
y = np.concatenate((y1,y2,y3),axis=0) #yukarıdan aşağıya birleştir
```

```python
dictionary = {"x":x,"y":y}
data = pd.DataFrame(dictionary)
```

```python
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
```

```python
# K means algoritmasınının göreceği data pointerlar
plt.scatter(x1,y1,color="black")
plt.scatter(x2,y2,color="black")
plt.scatter(x3,y3,color="black")
```

```python
# K değerini bulma
from sklearn.cluster import KMeans
wcss =[]
for k in range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,15),wcss)
plt.xlabel("K değeri")
plt.ylabel("wcss")
plt.show()
# Bu sonuçta K değeri 3 sayılır
```

```python
# K = 3

kmeans2 = KMeans(n_clusters=3)
cluster = kmeans2.fit_predict(data)
data["label"] =cluster
```

```python
plt.scatter(data.x[data.label ==0],data.y[data.label ==0],color="red")
plt.scatter(data.x[data.label ==1],data.y[data.label ==1],color="green")
plt.scatter(data.x[data.label ==2],data.y[data.label ==2],color="blue")
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color="yellow")
plt.show()
```
