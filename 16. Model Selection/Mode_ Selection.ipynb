```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y= iris.target
```

```python
#normalizasion
x = (x-np.min(x))/(np.max(x)-np.min(x))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# KNN model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# K FOLD CROSS VALİDATİON
from sklearn.model_selection import cross_val_score
accur = cross_val_score(estimator=knn,X=x_train,y=y_train,cv=10)
```

```python
# TEST DATAYA GÖRE TEST ETME
knn.fit(x_train,y_train)
print("test Accur:",knn.score(x_test,y_test))
```

GRİD SEARCH CROSS VALİDATİON
knn değerini belirlemek için kullanılır

```python
from sklearn.model_selection import GridSearchCV
grid ={"n_neighbors":np.arange(1,50)}
knn = KNeighborsClassifier()

knn_cv = GridSearchCV(knn,grid,cv=10)
knn_cv.fit(x,y)
print("K",knn_cv.best_params_)
print(knn_cv.best_score_)
```
