```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dataseti oluşturma  K değeri 3 np ile data oluşturma

x1 = np.random.normal(25,5,1000) # Gaussiyum dağılımı 25-5 arasında değerler üren gaussi değeri 100 olsun
y1 = np.random.normal(25,5,1000)

x2 = np.random.normal(55,5,1000) # Gaussiyum dağılımı 25-5 arasında değerler üren gaussi değeri 100 olsun
y2 = np.random.normal(60,5,1000)


x3 = np.random.normal(55,5,1000) # Gaussiyum dağılımı 25-5 arasında değerler üren gaussi değeri 100 olsun
y3 = np.random.normal(15,5,1000)
```

```python
x = np.concatenate((x1,x2,x3),axis=0) #yukarıdan aşağıya birleştir
y = np.concatenate((y1,y2,y3),axis=0) #yukarıdan aşağıya birleştir
```

```python
dictionary = {"x":x,"y":y}
data = pd.DataFrame(dictionary)
```

```python
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
merg =linkage(data,method="ward")
dendrogram(merg,leaf_rotation=90)
plt.xlabel("data point") 
plt.ylabel("oklid uzaklığı") 
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hiyerarsi = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
cluster = hiyerarsi.fit_predict(data)
data["label"] =cluster
```
